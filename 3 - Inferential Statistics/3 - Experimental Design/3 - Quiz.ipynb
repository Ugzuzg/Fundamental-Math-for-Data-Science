10. Fill in the code to simulate data for an A/B test where the probability of a “click” is 20% for version “A” of an ad and 30% for version “B”.

    ```python
    group_A = np.random.choice(['click', 'no click'],size=50,p=[___, ___])
    group_B = np.random.choice(['click', 'no click'],size=50,p=[___, ___])
    
    group = ['A']*50 + ['B']*50
    outcome = list(group_A) + list(group_B)
    sim_data = {"Version": group, "Outcome": outcome}
    ```

    - `.8`
    - `.7`
    - `.2`
    - `.3`

    <details>
        <summary>Answer</summary>
    
    ```python
    group_A = np.random.choice(['click', 'no click'],size=50,p=[.2, .8])
    group_B = np.random.choice(['click', 'no click'],size=50,p=[.3, .7])
    
    group = ['A']*50 + ['B']*50
    outcome = list(group_A) + list(group_B)
    sim_data = {"Version": group, "Outcome": outcome}
    ```
    </details>
