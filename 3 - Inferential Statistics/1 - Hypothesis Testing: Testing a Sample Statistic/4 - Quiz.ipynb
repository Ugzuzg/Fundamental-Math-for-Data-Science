6. A sample of 100 people who read an article took an average of 8 minutes to finish reading it. Suppose we have saved the amount of time that each of these 100 people took to read the article as a list named `reading_times`. We want to run a hypothesis test for this sample with the following null and alternative hypotheses:

    - Null hypothesis: On average, it will take readers 10 minutes to read the article
    - Alternative hypothesis: On average, it will take readers more or less than 10 minutes to read the article

    Fill in the code to run a one-sample t-test:

    ```python
    from scipy.stats import ttest_1samp
    tstat, pval = ttest_1samp(a = ___, popmean = ___)
    ```

    - `10`
    - `reading_times`
    - `8`

    <details>
        <summary>Answer</summary>
    
    ```python
    from scipy.stats import ttest_1samp
    tstat, pval = ttest_1samp(a = reading_times, popmean = 10)
    ```
    </details>

9. Out of a sample of 100 subscribers who saw an article, 15 people read it. Suppose we are running a hypothesis test for this sample with the following null and alternative hypotheses:

    - Null hypothesis: 10% of subscribers who see an article will read it
    - Alternative hypothesis: More than 10% of subscribers who see an article will read it

    Fill in the code to run a binomial hypothesis test:

    ```python
    from scipy.stats import binom_test
    p_value_2sided = binom_test(x = ___, n = ___, p = ___, alternative = ___)
    ```

    - `100`
    - `'two-sided'`
    - `'less'`
    - `0.1`
    - `15`
    - `'greater'`

    <details>
        <summary>Answer</summary>
    
    ```python
    from scipy.stats import binom_test
    p_value_2sided = binom_test(x = 15, n = 100, p = 0.1, alternative = 'greater')
    ```
    </details>
