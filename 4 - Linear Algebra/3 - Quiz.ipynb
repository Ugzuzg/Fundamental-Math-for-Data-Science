    $$

    $$
    C = BA
    $$

    ```python
    import numpy as np
    A = np.array([2, 1]) ___ np.array([-4, 2])
    B = 2*np.array([___]) + np.array([[0,2],[-1,-1]])
    C = ___@___
    ```

    - `-`
    - `[3,-1],[1,-2]`
    - `[3, 1],[-1,-2]`
    - `+`

    <details>
