1. Suppose that the average height of an adult is 67 inches, with a standard deviation of 3 inches.

    Fill in the code below to correctly answer the following question: What is the probability that a randomly sampled group of 10 adults has an average height that is less than or equal to 65 inches?

    ```python
    stats.norm.___(___, ___, ___)
    ```

    - `3/(10**.5)`
    - `3`
    - `cdf`
    - `65`
    - `3/10`
    - `67`
    - `pdf`

    <details>
        <summary>Answer</summary>
    
    ```python
    stats.norm.cdf(65, 67, 3/(10**.5))
    ```
    </details>
