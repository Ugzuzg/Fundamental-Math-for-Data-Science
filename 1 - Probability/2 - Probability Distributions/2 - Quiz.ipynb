3. Fill in the correct vocabulary in the following statements:

    A ___ random variable has a countable number of possible values. These random variables are often observed when ___ something such as the number on a die roll or number of people in a room.
    
    A ___ random variable has an uncountable number of possible values. These random variables are often ___ of some kind. For example time, distance, or temperature.

    - `discrete`
    - `counting`
    - `measurements`
    - `continuous`
    
    <details>
        <summary>Answer</summary>
    
    A discrete random variable has a countable number of possible values. These random variables are often observed when counting something such as the number on a die roll or number of people in a room.
    
    A continuous random variable has an uncountable number of possible values. These random variables are often measurements of some kind. For example time, distance, or temperature.
    </details>

7. Fill in the blanks to create code that will calculate the probability of observing more than 3 and fewer than 7 heads from 10 fair coin flips.

    ```python
    import scipy.stats as st
    
    st.binom.cdf(___, n=10, p=___) - st.binom.cdf(___, n=10, p=0.5)
    ```

    - `4`
    - `0.5`
    - `7`
    - `3`
    - `6`

    <details>
        <summary>Answer</summary>

    ```python
    st.binom.cdf(6, n=10, p=0.5) - st.binom.cdf(3, n=10, p=0.5)
    ```
    </details>

10. Fill in the code to produce the probability of observing exactly 3 heads in 10 fair coin flips.

    ```python
    import scipy.stats as st

    ```

    - `3`
    - `.5`
    - `cdf`

    - `pmf`
